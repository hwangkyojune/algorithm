```python
def insertion(A):
    for i in range(len(A)-1):
        j = i
        while A[j] > A[j+1] and j >= 0:
            A[j],A[j+1] = A[j+1],A[j]
            j -= 1
    return A
```

```python
def get_max(A,i):
    cm=index=0
    for j in range(0,i+1):
        if cm < A[j]:
            cm = A[j]
            index = j
    A[i],A[index] = A[index],A[i]
```

```python
def selection(A):
    for i in range(len(A)-1,0,-1):
        get_max(A,i)
        
    return A
```

```python
A=[6,3,7,2,5]
print(selection(A))
```

```python
def quick_select1(A): #stable하지만 not inplace한 quick select
    
    S = []
    L = []
    M = []
    
    if len(A) <=1: return A
    
    m = (len(A) -1) //2
    
    for i in range(0,len(A)):
        
        if A[i] < A[m] :
                S.append(A[i])
                
        elif A[i] > A[m] :
                L.append(A[i])
                
        else :
                M.append(A[m])
                
    return quick_select(S) + M + quick_select(L)
            
```

```python
def quick_sort(A,first,last):
    
    if first >= last : return 
    
    pivot = A[first]
    left,right = first+1, last
    
    while left <= right:
        while left <= last and A[left] <= pivot:
            left+=1
        while right > first and A[right] >= pivot:
            right-=1
        if left <= right:
            A[left],A[right] = A[right],A[left]
            left +=1
            right -=1
            
    A[first],A[right] = A[right],A[first]
    quick_sort(A,first,right-1)
    quick_sort(A,right+1,last)
```

```python
def merge_two_sorted(A,first,last):
    global Mc
    global Ms
    B = []
    fi,se = first, (first+last)//2 + 1
    
    while (fi <= (first+last)//2) and (se <= last):
        Mc += 1
        if A[fi] <= A[se]:
            B.append(A[fi])
            fi += 1
        else :
            B.append(A[se])
            se += 1
    
    for i in range(fi,(first+last)//2+1):
        B.append(A[i])
    
    for i in range(se,last+1):
        B.append(A[i])
    
    for k in range(first,last+1):
        Ms += 1
        A[k] = B[k-first]
    #같은 수를 가지고 있더라도 어쨋든 대입을 하기 때문에 교환 횟수에 더해줌
```

```python
def merge_sort(A,first,last):
    
    if first >= last : return
    
    merge_sort(A,first,(first+last)//2)
    merge_sort(A,(first+last)//2+1,last)
    merge_two_sorted(A,first,last)
```

```python
def merge_sorted2(A,first,m,last):
    B = []
    fi,se = first, m+1
    
    while fi <= m and se <= last:
        if A[fi] <= A[se]:
            B.append(A[fi])
            fi += 1
        else :
            B.append(A[se])
            se += 1
    
    for k in range(fi, m+1):
        B.append(A[fi])
    for k in range(se,last+1):
        B.append(A[se])
    
    for i in range(first,last+1):
        A[i] = B[i-first]
```

```python
def merge_sort3(A,first,last):
    
    if first >= last : return
    
    merge_sort3(A,first,first+(last-first)//3)
    merge_sort3(A,first+(last-first)//3+1,first + 2*(last-first)//3)
    merge_sort3(A,first+ 2*(last-first)//3 + 1,last)
                
    merge_sorted2(A,first,first+(last-first)//3,first + 2*(last-first)//3)
    merge_sorted2(A,first,first+ 2*(last-first)//3,last)
```
